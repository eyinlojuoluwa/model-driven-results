```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


# Define the GitHub API URL for searching repositories
api_url = "https://api.github.com/search/repositories"


def parse_number_with_suffix(number_str):
    # Remove any commas in the number string and convert it to lowercase
    number_str = number_str.replace(',', '').lower()
    if 'k' in number_str:
        return int(float(re.sub(r'k$', '', number_str)) * 1000)
    elif 'm' in number_str:
        return int(float(re.sub(r'm$', '', number_str)) * 1000000)
    else:
        return int(number_str)

# Define your search parameters
params = {
    "q": "cookiecutter",
    "type": "Repositories",
    "sort": "stars",
    "order": "desc",
    "per_page": 100,
}

# GitHub API token for authentication
headers = {"Authorization": "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"}

# Send a GET request to the GitHub API
response = requests.get(api_url, params=params, headers=headers)

if response.status_code == 200:
    data = response.json()
    items = data.get("items", [])

    repo_names = []
    stars = []
    contributors = []
    commits = []
    issues = []

    for repo in items:
        repo_names.append(repo["full_name"])
        stars.append(repo["stargazers_count"])

        # Parse the HTML of the repository page to get more details
        repo_url = repo["html_url"]
        repo_response = requests.get(repo_url, headers=headers)
        if repo_response.status_code == 200:
            repo_soup = BeautifulSoup(repo_response.text, "html.parser")

            # Get contributors count from the contributors page
            contributors_url = repo_url + "/graphs/contributors"
            contributors_response = requests.get(contributors_url, headers=headers)
            if contributors_response.status_code == 200:
                contributors_soup = BeautifulSoup(contributors_response.text, "html.parser")

                # Find the contributors count
                contributors_count_str = contributors_soup.find("h2", class_="h4 text-normal").find("span").text
                contributors_count = parse_number_with_suffix(contributors_count_str)
                contributors.append(contributors_count)
            else:
                contributors.append(None)

            # Get commits count from the commit-activity page
            commit_activity_url = repo_url + "/graphs/commit-activity"
            commit_activity_response = requests.get(commit_activity_url, headers=headers)
            if commit_activity_response.status_code == 200:
                commit_activity_soup = BeautifulSoup(commit_activity_response.text, "html.parser")
                commits_count_str = commit_activity_soup.find("h2", class_="h4 text-normal").find("span").text
                commits_count = parse_number_with_suffix(commits_count_str)
                commits.append(commits_count)
            else:
                commits.append(None)

            # Get issues count
            issues_url = repo_url + "/issues?q=is%3Aall+is%3Aissue"
            issues_response = requests.get(issues_url, headers=headers)
            if issues_response.status_code == 200:
                issues_soup = BeautifulSoup(issues_response.text, "html.parser")
                issues_count_str = issues_soup.find("span", class_="Counter").text
                issues_count = parse_number_with_suffix(issues_count_str)
                issues.append(issues_count)
            else:
                issues.append(None)
        else:
            contributors.append(None)
            commits.append(None)
            issues.append(None)

    # Create a DataFrame
    data = {
        "repo_name": repo_names,
        "number_of_stars": stars,
        "number_of_contributors": contributors,
        "number_of_commits": commits,
        "number_of_issues": issues,
    }
    df = pd.DataFrame(data)

    # Display the first 100 repositories
    df.to_csv("github_doc.csv", index=None)
    #print(df)
else:
    print("Failed to retrieve data from the GitHub API. Status code:", response.status_code)
```
